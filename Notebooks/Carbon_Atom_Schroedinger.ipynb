```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D

# ==========================================
# OUTPUT
# ==========================================

OUTPUT_VIDEO = "../Video_Animation/carbon_atom_schrodinger.mp4"

# ==========================================
# SETTINGS
# ==========================================

fps = 30

frames_1s = 250
frames_2s2p = 250

total_frames = frames_1s + frames_2s2p

limit = 12

# ==========================================
# FIGURE
# ==========================================

fig = plt.figure(figsize=(10,10))

fig.patch.set_facecolor("black")

ax = fig.add_subplot(111, projection='3d')

ax.set_facecolor("black")

# ==========================================
# POSITION STORAGE
# ==========================================

positions_1s = []
positions_2s = []
positions_2p = []

# ==========================================
# 1s SCHRÖDINGER ORBITAL
# ==========================================

def psi_1s(r):

    return np.exp(-r)

# ==========================================
# 2s SCHRÖDINGER ORBITAL
# ==========================================

def psi_2s(r):

    return (2 - r) * np.exp(-r / 2)

# ==========================================
# 2p SCHRÖDINGER ORBITAL
# ==========================================

def psi_2p(r, theta):

    angular = np.cos(theta)

    radial = r * np.exp(-r / 2)

    return radial * angular

# ==========================================
# SAMPLING
# ==========================================

def sample_1s():

    while True:

        x = np.random.uniform(-6, 6)
        y = np.random.uniform(-6, 6)
        z = np.random.uniform(-6, 6)

        r = np.sqrt(x**2 + y**2 + z**2)

        probability = np.abs(psi_1s(r))**2

        if np.random.rand() < probability:

            return x, y, z

# ==========================================

def sample_2s():

    while True:

        x = np.random.uniform(-10, 10)
        y = np.random.uniform(-10, 10)
        z = np.random.uniform(-10, 10)

        r = np.sqrt(x**2 + y**2 + z**2)

        probability = np.abs(psi_2s(r))**2

        probability = probability / 4

        if np.random.rand() < probability:

            return x, y, z

# ==========================================

def sample_2p():

    while True:

        x = np.random.uniform(-10, 10)
        y = np.random.uniform(-10, 10)
        z = np.random.uniform(-10, 10)

        r = np.sqrt(x**2 + y**2 + z**2)

        if r == 0:
            continue

        theta = np.arccos(z / r)

        probability = np.abs(psi_2p(r, theta))**2

        probability = probability / 5

        if np.random.rand() < probability:

            return x, y, z

# ==========================================
# UPDATE
# ==========================================

def update(frame):

    ax.clear()

    ax.set_facecolor("black")

    # ======================================
    # KERN
    # ======================================

    ax.scatter(
        0,
        0,
        0,
        s=1200,
        c="orange",
        alpha=1.0
    )

    # ======================================
    # PHASE 1
    # ======================================

    if frame < frames_1s:

        for _ in range(2):

            positions_1s.append(
                sample_1s()
            )

    # ======================================
    # PHASE 2
    # ======================================

    else:

        for _ in range(2):

            positions_2s.append(
                sample_2s()
            )

        for _ in range(2):

            positions_2p.append(
                sample_2p()
            )

    # ======================================
    # DRAW 1s
    # ======================================

    if len(positions_1s) > 0:

        p1 = np.array(positions_1s)

        ax.scatter(
            p1[:,0],
            p1[:,1],
            p1[:,2],
            s=6,
            c="cyan",
            alpha=0.05
        )

    # ======================================
    # DRAW 2s
    # ======================================

    if len(positions_2s) > 0:

        p2 = np.array(positions_2s)

        ax.scatter(
            p2[:,0],
            p2[:,1],
            p2[:,2],
            s=4,
            c="deepskyblue",
            alpha=0.03
        )

    # ======================================
    # DRAW 2p
    # ======================================

    if len(positions_2p) > 0:

        p3 = np.array(positions_2p)

        ax.scatter(
            p3[:,0],
            p3[:,1],
            p3[:,2],
            s=4,
            c="white",
            alpha=0.03
        )

    # ======================================
    # STYLE
    # ======================================

    ax.set_xlim(-limit, limit)
    ax.set_ylim(-limit, limit)
    ax.set_zlim(-limit, limit)

    ax.grid(False)

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])

    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_zlabel("")

    ax.set_title(
        "Carbon Atom Schrödinger Simulation",
        color="white"
    )

# ==========================================
# CREATE ANIMATION
# ==========================================

ani = FuncAnimation(
    fig,
    update,
    frames=total_frames,
    interval=1000 / fps
)

# ==========================================
# SAVE VIDEO
# ==========================================

print("Schrödinger-Video wird erstellt...")

ani.save(
    OUTPUT_VIDEO,
    writer="ffmpeg",
    fps=fps
)

print("Video gespeichert unter:")
print(OUTPUT_VIDEO)

plt.close()
```
