```python
# ==========================================
# IBM QUANTUM + H2 MOLEKÜL
# SIMULATOR + ECHTE HARDWARE
# ==========================================

# ==========================================
# IMPORTS
# ==========================================

from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

# ==========================================
# IBM SERVICE VERBINDUNG
# ==========================================

service = QiskitRuntimeService()

print("Verfügbare Backends:")
print(service.backends())

# ==========================================
# BACKEND AUSWÄHLEN
# ==========================================

# ------------------------------------------
# OPTION 1:
# IBM CLOUD SIMULATOR
# ------------------------------------------

backend = service.backend("ibm_kingston")

# ------------------------------------------
# SPÄTER FÜR ECHTE HARDWARE:
#
# backend = service.backend("ibm_fez")
#
# ODER:
#
# backend = service.backend("ibm_marrakesh")
# ------------------------------------------

print("\nAktuelles Backend:")
print(backend)

# ==========================================
# EINFACHE H2-ÄHNLICHE QUANTENSCHALTUNG
# ==========================================

# 2 Qubits
qc = QuantumCircuit(2)

# ==========================================
# QUANTENÜBERLAGERUNG
# ==========================================

qc.h(0)

# ==========================================
# VERSCHRÄNKUNG
# ==========================================

qc.cx(0, 1)

# ==========================================
# MESSUNG
# ==========================================

qc.measure_all()

print("\nQuantum Circuit:")
print(qc)

# ==========================================
# SCHALTUNG OPTIMIEREN
# ==========================================

pm = generate_preset_pass_manager(
    backend=backend,
    optimization_level=1
)

isa_circuit = pm.run(qc)

# ==========================================
# MODERNER IBM QUANTUM SAMPLER
# ==========================================

from qiskit_ibm_runtime import SamplerV2 as Sampler

# ==========================================
# SAMPLER ERSTELLEN
# ==========================================

sampler = Sampler(mode=backend)

# ==========================================
# JOB ABSENDEN
# ==========================================

print("\nJob wird abgesendet...")

job = sampler.run(
    [isa_circuit],
    shots=2048
)

# ==========================================
# JOB ID
# ==========================================

print("\nJob-ID:")
print(job.job_id())

# ==========================================
# STATUS ABFRAGEN
# ==========================================

print("\nBitte warten...")

import time

while not job.done():

    print("Status:", job.status())

    time.sleep(10)

print("\nJob abgeschlossen.")

# ==========================================
# ERGEBNIS HOLEN
# ==========================================

result = job.result()

# ==========================================
# COUNTS EXTRAHIEREN
# ==========================================

pub_result = result[0]

counts = pub_result.data.meas.get_counts()

# ==========================================
# AUSGABE
# ==========================================

print("\nMESSERGEBNISSE:")
print(counts)

# ==========================================
# WAHRSCHEINLICHKEITEN
# ==========================================

print("\nWAHRSCHEINLICHKEITEN:")

for state, count in counts.items():

    probability = count / total_shots

    print(f"{state}: {probability:.4f}")

# ==========================================
# WAHRSCHEINLICHKEITEN BERECHNEN
# ==========================================

total_shots = sum(counts.values())

print("\nWAHRSCHEINLICHKEITEN:")

for state, count in counts.items():

    probability = count / total_shots

    print(f"Zustand {state}: {probability:.4f}")

# ==========================================
# INTERPRETATION
# ==========================================

print("\nINTERPRETATION:")

print("""
Die Zustände zeigen die Wahrscheinlichkeits-
verteilung der verschränkten Qubits.

Idealerweise dominieren:
00 und 11

Das entspricht einem Bell-Zustand
(quantum entanglement).

Auf echter Hardware entstehen zusätzlich:
01 und 10

durch:
- Noise
- Gatefehler
- Dekohärenz
- Messfehler
""")
```
